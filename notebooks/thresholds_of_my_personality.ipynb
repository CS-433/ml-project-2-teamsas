```python
import pandas as pd
```

```python
df = pd.read_csv('./external_datasets/my_personality/my_personality.csv', encoding="ISO-8859-1",)
df.head()
```

```python
cont_vars = ['sEXT', 'sNEU', 'sAGR', 'sCON', 'sOPN']
discrete_vars = ['cEXT', 'cNEU', 'cAGR', 'cCON', 'cOPN']
```

```python
for var in cont_vars:
    df[var] = (df[var] - df[var].min()) / (df[var].max() - df[var].min())

for var in discrete_vars:
    df[var] = df[var].apply(lambda x: True if x == 'y' else False)

df.head()
```

```python
thresholds = []

for cont_var, discrete_var in zip(cont_vars, discrete_vars):
    max_score_for_negative = df[df[discrete_var] == False][cont_var].max()
    min_score_for_positive = df[df[discrete_var] == True][cont_var].min()
    threshold = (max_score_for_negative + min_score_for_positive) / 2
    thresholds.append(threshold)
    print(f"{cont_var} -> {discrete_var}: {max_score_for_negative:.4f} - {min_score_for_positive:.4f} => {threshold:.4f}")

print()
print(thresholds)
```
