# EDA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
data_df = pd.read_excel("../../data/dataset.xlsx")
data_df = data_df.dropna()
data_df = data_df.reset_index(drop=True)
data_df = data_df.drop(columns=["Unnamed: 0"])
data_df.head()
```

```python
my_personality_df = pd.read_csv(
    "../../external_datasets/my_personality/my_personality.csv",
    encoding="ISO-8859-1",
)
my_personality_df = my_personality_df.dropna()
my_personality_df.head()
```

## Target Variables

### Main Dataset

```python
targets = [
    "hones16",
    "emoti16",
    "extra16",
    "agree16",
    "consc16",
    "openn16",
]

labels = [
    "Honesty",
    "Emotional Stability",
    "Extroversion",
    "Agreeableness",
    "Conscientiousness",
    "Openness",
]

data_vars = data_df[targets].copy(deep=True)
data_vars.columns = labels
data_vars["Dataset"] = "IDIAP"
data_vars
```

```python
my_personality_targets = ["sEXT", "sNEU", "sAGR", "sCON", "sOPN"]

my_personality_labels = [
    "Extroversion",
    "Emotional Stability",
    "Agreeableness",
    "Conscientiousness",
    "Openness",
]

my_personality_vars = my_personality_df[my_personality_targets].copy(deep=True)
my_personality_vars.columns = my_personality_labels
my_personality_vars.loc[:, "Dataset"] = "My Personality"
my_personality_vars
```

```python
plt.figure()
sns.violinplot(data=data_df[targets], fill=False)
plt.xticks(range(6), labels)
plt.xticks(rotation=15)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.ylabel("Score")
plt.tight_layout()
plt.savefig("targets_violin.pdf", format="pdf")
plt.show()
```

```python
shared_labels = [
    "Extroversion",
    "Emotional Stability",
    "Agreeableness",
    "Conscientiousness",
    "Openness",
    "Honesty",
]

joint_vars = pd.concat([data_vars, my_personality_vars], axis=0)[shared_labels + ["Dataset"]]
joint_vars = joint_vars.reset_index(drop=True)
joint_vars = joint_vars.melt(id_vars=["Dataset"], var_name="Trait", value_name="Score")

plt.figure()
sns.violinplot(data=joint_vars, x="Trait", y="Score", hue="Dataset", split=True, gap=.1, inner="quart", fill=False, linewidth=2.5)
plt.xticks(range(6), shared_labels)
plt.xticks(rotation=30)
plt.xticks(fontsize=11)
plt.xlabel("")
plt.yticks(fontsize=11)
plt.ylim(0.5, 5.5)
plt.legend(title="Dataset", fontsize=9, loc="lower right", ncol=2)
plt.tight_layout()
plt.savefig("both_targets_violin.pdf", format="pdf")
plt.show()
```
